```python
role_tags_true = pd.read_csv(f'{data_folder}/role_tags_true.csv', header=0) 
role_tags_true_17985 = role_tags_true[role_tags_true.activity_id == 17985].reset_index()
form_tags_true_17985 = role_tags_true_17985[role_tags_true_17985.columns[1:-2]].drop_duplicates().reset_index(drop=True)
```

```python
def add_formcp(records):
  indexes = records.index
  for i in indexes[:-1]:
    records.loc[i, "form_cp"] = ((records.loc[i, "formation"] != records.loc[i+1, "formation"]) | (records.loc[i, "session"] != records.loc[i+1, "session"])) * (records.loc[i, "end_time"])
  records.loc[indexes[-1], 'form_cp'] = records.loc[indexes[-1], 'end_time']
```

```python
def dtw_metric(gt, pred):
  # print("GT :", gt)
  # print("Pred : ", pred)
  alignement = dtw(gt, pred)
  return alignement.distance

def directed_hausdorff(u, v):
  min_values = []
  for i in u:
    dist = [abs(i - j) for j in v]
    min_dist = min(dist)
    min_values.append(min_dist)
  return max(min_values)


def hausdorff(gt, pred):
  return max(directed_hausdorff(gt, pred), directed_hausdorff(pred, gt))


def formation_evaluation(datafolder, groundtruth, mode="gseg_avg", metric=hausdorff, distance="", verbose=False):
  total_metric = 0
  pred_form = pd.read_pickle(f"{datafolder}/{mode}/form/17985{distance}.pkl")

  for session in np.unique(pred_form.session):
    
    session_idx = pred_form[pred_form.session == session].index
    pred_session_start_dt =  min(pred_form.loc[session_idx, "start_dt"])
    pred_form.loc[session_idx, "form_cp"] = pred_form.loc[session_idx, "end_dt"] - pred_session_start_dt
    pred_form.loc[session_idx, "form_cp"] = pred_form.loc[session_idx, 'form_cp'].apply(lambda x: np.ceil(x.seconds / 60))

    gt_cpd = groundtruth[groundtruth.session == session]["form_cp"].unique().tolist()
    if 0 in gt_cpd:
      gt_cpd.remove(0)
    gt_cpd = [0] + gt_cpd

    pred_cpd = [0] + pred_form.loc[session_idx]["form_cp"].unique().tolist()
    if verbose:
      print(f"GT : {gt_cpd}")
      print(f"PRED : {pred_cpd}")
    metric_value = metric(gt_cpd, pred_cpd)
    print("Session {} : metric = {}".format(session, metric_value))
    total_metric += metric_value
  return total_metric
```

```python
cdp_modes = ['gseg_avg', 'gseg_union', 'kernel_linear', 'kernel_rbf', 'kernel_cosine', 'rank']
for mode in cdp_modes:
  print("\nChange-point-detection algo : {}".format(mode))
  result = formation_evaluation(data_folder, form_tags_true_17985, mode)
  print("Global error : {}".format(result))
```

### Role periods

```python
def add_formation_number(gt):
  gt["form_period"] = 1
  for i in gt.index[1:]:
    if gt.loc[i - 1, "formation"] != gt.loc[i, "formation"] or gt.loc[i - 1, "session"] != gt.loc[i, "session"]:
      gt.loc[i, "form_period"] = gt.loc[i - 1, "form_period"] + 1
    else :
      gt.loc[i, "form_period"] = gt.loc[i - 1, "form_period"]
```

```python
add_formation_number(role_tags_true_17985)
```

```python
def get_rolecpd_list_pred(df, start_session):
  cpd_list = [0]
  if len(df) == 1:
    return cpd_list + [np.ceil((df.iloc[0]["end_dt"] - start_session).seconds / 60)]

  for i in range(1, len(df)):
    if df.iloc[i]["base_role"] != df.iloc[i - 1]["base_role"]:
      cpd_list.append(np.ceil((df.iloc[i]["start_dt"] - start_session).seconds / 60))

  cpd_list.append(df.iloc[-1]["end_dt"])
  return cpd_list

def get_rolecpd_list_gt(df):
  cpd_list = [0]
  if len(df) == 1:
    return cpd_list + [df.iloc[0]["end_time"]]
  
  for i in range(1, len(df)):
    if df.iloc[i]["aligned_role"] != df.iloc[i - 1]["aligned_role"]:
      cpd_list.append(df.iloc[i]["start_time"])

  cpd_list.append(df.iloc[-1]["end_time"])
  return cpd_list
```

```python
def role_evaluation(data_folder, groundtruth, mode="gseg_avg", metric=hausdorff):
  total_metric = 0
  pred_role = pd.read_csv(f"{data_folder}/gseg_avg/{mode}/role/17985.csv")
  pred_role["end_dt"] = pred_role["end_dt"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
  pred_role["start_dt"] = pred_role["start_dt"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))

  for session in np.unique(groundtruth.session):
    metric_session = 0
    
    gt_session = groundtruth[groundtruth.session == session]
    pred_session = pred_role[pred_role.session == session]
    start_session = min(pred_session["start_dt"])

    for form in gt_session["form_period"].unique():
      gt_byform = gt_session[gt_session['form_period'] == form]
      pred_byform = pred_session[pred_session["form_period"] == form] # on sait que si la formcpd_type = avg_seg la formation est bien predite


      for player in gt_byform["squad_num"].unique():
        gt_player = gt_byform[gt_byform["squad_num"] == player]

        if (gt_player["aligned_role"] == "GK").all():
          continue
        pred_player = pred_byform[pred_byform["squad_num"] == player]

        gt_cpd = get_rolecpd_list_gt(gt_player)
        pred_cpd = get_rolecpd_list_pred(pred_player, start_session)

        metric_value = metric(gt_cpd, gt_cpd)
        metric_session += metric_value

    print("Session {} : metric = {}".format(session, metric_session))
    total_metric += metric_session
    
  return total_metric
```

```python
rolescdp_modes = ['gseg_avg', 'gseg_union']
for mode in rolescdp_modes:
  print("\nChange-point-detection algo : {}".format(mode))
  result = role_evaluation(data_folder, role_tags_true_17985, mode)
  print("Global error : {}".format(result))
```
